# init matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
y = np.exp(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
# plt.show()
```

```python
plt.plot(x, y, '--')
```

```python
plt.plot(x, y, 'r^')
```

```python
plt.plot(x, y, 'go')
```

```python
plt.plot(x, y, 'bv')
```

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y)
plt.xlabel('Effort')
plt.ylabel('Earning')
plt.title('This is how your effort earns')
```

```python
plt.plot(x, y, label='Earning with effort')
plt.legend()
plt.xticks(np.arange(-3, 4, 0.5))
plt.yticks([0, 5, 10, 20])
plt.show()
```

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x) * 2
plt.plot(x, y1, label='first')
plt.plot(x, y2, label='second')
plt.legend()
plt.axis('off')
plt.show()
```

# subplot

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = x * x

plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
```

```python
# Object-oriented plotting
fig = plt.figure()
type(fig)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, y1)
ax2.plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1)
axes[1].plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
print(axes.shape)
```

```python
# important
fig, axes = plt.subplots(nrows=3, ncols=3)
axes[1, 2].plot(x, y2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y1, label='something')
axes[1].plot(x, y1)
axes[0].set_xlabel('xlabel1')
axes[0].set_ylabel('xlabel2')
axes[0].set_title('plot title')
axes[0].set_xticks([-3, -2, -1, 3])
axes[0].set_yticks([0, 10, 20])
axes[0].legend()
axes[1].axis('off')
```
