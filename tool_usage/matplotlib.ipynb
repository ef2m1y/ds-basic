# subplot

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = x * x

plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
```

```python
# Object-oriented plotting
fig = plt.figure()
type(fig)
```

# saving

```python
x = np.linspace(-3, 3, 10)
y1 = np.exp(x)
y2 = np.exp(x) * 2

fig, axes = plt.subplots()
axes.plot(x, y1)
```

```python
fig, axes = plt.subplots(figsize=(1, 1), dpi=100)
axes.plot(x, y1)
```

```python
fig, axes = plt.subplots(figsize=(10, 2))
axes.plot(x, y1)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
axes[0,].plot(x, y1)
axes[1,].plot(x, x * x)

axes[0,].set_title('first')
axes[1,].set_title('second')

axes[0,].set_xlabel('x')
axes[0,].set_ylabel('y')
axes[1,].set_xlabel('x')
axes[1,].set_ylabel('y')
# png
fig.savefig('savefig_sample.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
axes[0,].plot(x, y1)
axes[1,].plot(x, x * x)

axes[0,].set_title('first')
axes[1,].set_title('second')

axes[0,].set_xlabel('x')
axes[0,].set_ylabel('y')
axes[1,].set_xlabel('x')
axes[1,].set_ylabel('y')

plt.tight_layout()

fig.savefig('savefig_sample.png')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
axes.legend()
```

```python
fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
# default: loc='best'
axes.legend(loc='center left')
```

pdf saving

```python
from matplotlib.backends.backend_pdf import PdfPages
# pdf instance
pdf = PdfPages('savefig_sample.pdf')

fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
# default: loc='best' ('best' == 0)
axes.legend(loc=0)

pdf.savefig(fig)

pdf.close()
```

```python
pdf = PdfPages('savemultifig_sample.pdf')
# axes->axes1->PDFi->same...
for i in range(0, 10):
    fig, axes = plt.subplots()
    axes.plot(x, y1 + x * i)
    axes.set_title('ID:#{}'.format(i))
    pdf.savefig(fig)
    
pdf.close()

# ctrl + F: "#4"
```
